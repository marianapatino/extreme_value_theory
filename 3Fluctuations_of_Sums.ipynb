Verifying the Central Limit Theorem


*   drawing sample from exponential distribution with $\lambda$ = llambda = 1/scale

*   considering 6 cases with varying number of r.v.s

*   more draws -> resembles the normal even better (CLT)

*   more r.v.s  -> smaller std in approximated normal form

```python
#drawing sample from exponential distribution with lambda = llambda = 1/scale
#considering 6 cases with varying number of r.v.s
#more draws -> resembles the normal even better (CLT)
#more r.v.s  -> smaller std in approximated normal form

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import math

f = plt.figure()
fig, ((ax1,ax2,ax3),(ax4,ax5,ax6)) = plt.subplots(2,3,figsize=(14,12))
for size, ax in zip([2**i for i in range(1,7)],[ax1,ax2,ax3,ax4,ax5,ax6]):
    means = []
    draws = 10000
    llambda = 1
    for _ in range(draws):
        sample = np.random.exponential(scale=1/llambda,size=size) 
        means.append(sample.mean())
   
    neu = (np.asarray(means)-1/llambda)/((1/llambda)/math.sqrt(size))
    x = np.linspace(- 4, 4,100)
    ax.plot(x, norm.pdf(x, 0, 1),label='$\mathcal{N}(0,1)$',color="darkorange")    
    ax.hist(neu,bins=100,density=True,label='sample mean (Exp(1))',color="lavender")
    ax.set_xlim(-4,4)
    ax.title.set_text(f'n = {size}')
    ax.set_ylim(0,0.65)
    #ax.text(.2,.9,f'n = {size}',horizontalalignment='right', transform=ax.transAxes)
    ax.set_xlabel('x')
    #ax.legend(loc='lower left', bbox_to_anchor= (0.0, 1.01), ncol=2, 
            #borderaxespad=0, frameon=False)
    ax.legend()



    '''print('Mean:', np.asarray(means).mean())
    print('Std:', np.asarray(means).std())'''

plt.savefig(filepath+"CLT_Exponential&Normal.png",bbox_inches = 'tight')
plt.show()
```
