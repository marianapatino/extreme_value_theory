1. Poisson distribution

```
from scipy.stats import poisson
import numpy as np 
import matplotlib.pyplot as plt

x = np.arange(0,20,1)
mycolors = {1: 'orange', 5: 'blue', 10: 'green'}

f = plt.figure()
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(13,4))

ax1.set_xlim(-1,20)
for i in 1,5,10:
  ax1.plot(x, poisson.pmf(x,i),'-',color=mycolors[i])
  ax1.plot(x, poisson.pmf(x,i),'o',color=mycolors[i],
           label=f'$\lambda$ = {i}')
ax1.set_xlabel("x")
#ax1.set_ylabel("P(X=x)")
ax1.title.set_text("pmf")
ax1.legend()

for i in 1,5,10:
  ax2.plot(x,poisson.cdf(x,i),color=mycolors[i],)
  ax2.plot(x, poisson.cdf(x,i),'o',color=mycolors[i],
           label=f'$\lambda$ = {i}')
ax2.set_xlabel("x")
#ax2.set_ylabel("P(X$\leq$x)")
ax2.title.set_text("cdf")
ax2.legend()

plt.savefig(filepath+"Poisson_pmf&cdf.png", bbox_inches = 'tight')
plt.show()
```
